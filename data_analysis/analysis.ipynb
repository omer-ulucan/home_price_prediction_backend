The following functions will help to find :
average price


```python
def avg_price(df):
    return round(df["price"].sum() / len(df), 2)

def avg_price_by_area(df, area=None):
    if area:  # If specific area is provided
        if area in df["area"].values:
            return round(df[df["area"] == area]["price"].mean(), 2)
        else:
            raise KeyError(f"Area {area} is not in the DataFrame.")
    else:  # If no area is provided, return average price by area
        return df.groupby("area")["price"].mean().round(2)

print(avg_price_by_area(housing, 7420))
```

furnished dataframe

```python
def is_furnished(df):
    furnished_df = pd.concat([df[df["furnishingstatus"] == "furnished"], df[df["furnishingstatus"] == "semi-furnished"]])
    return furnished_df
    

furnished_df = is_furnished(housing)
furnished_df.head()
```
